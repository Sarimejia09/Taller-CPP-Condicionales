```python
# Importamos la librería necesaria para trabajar con modelos de aprendizaje automático
from sklearn.linear_model import LogisticRegression

# Importamos numpy para manejar arreglos numéricos
import numpy as np

# Creamos los datos de entrada (X)
# Cada valor representa la edad de una persona
X = np.array([[18], [22], [25], [35], [45], [55]])

# Creamos las etiquetas (y)
# 0 = no compra, 1 = compra
y = np.array([0, 0, 0, 1, 1, 1])

# Creamos el modelo de aprendizaje supervisado
modelo = LogisticRegression()

# Entrenamos el modelo con los datos (aprende la relación entre edad y compra)
modelo.fit(X, y)

# Creamos un nuevo dato para predecir (una persona de 30 años)
nueva_persona = np.array([[30]])

# Usamos el modelo para hacer una predicción
prediccion = modelo.predict(nueva_persona)

# Mostramos el resultado en pantalla
print("¿Comprará? (0 = No, 1 = Sí):", prediccion[0])
```

```python
# Importamos la librería KMeans para hacer agrupamiento (clustering)
from sklearn.cluster import KMeans

# Importamos numpy para manejar datos numéricos en forma de arreglos
import numpy as np

# Creamos los datos (X)
# Cada fila representa una persona con dos características:
# [edad, dinero gastado]
X = np.array([
    [18, 100],
    [20, 150],
    [22, 130],
    [40, 800],
    [42, 850],
    [45, 900]
])

# Definimos el modelo KMeans
# n_clusters=2 significa que queremos formar 2 grupos
modelo = KMeans(n_clusters=2)

# Entrenamos el modelo con los datos
# Aquí el algoritmo intenta encontrar grupos similares
modelo.fit(X)

# Obtenemos a qué grupo pertenece cada dato
grupos = modelo.labels_

# Mostramos los grupos en pantalla
print("Grupo de cada persona:", grupos)

# Creamos un nuevo dato (persona nueva)
nueva_persona = np.array([[25, 200]])

# Predecimos a qué grupo pertenece esta nueva persona
grupo_nuevo = modelo.predict(nueva_persona)

# Mostramos el resultado
print("La nueva persona pertenece al grupo:", grupo_nuevo[0])
```

```python
# Importamos numpy para trabajar con matrices y números
import numpy as np

# Definimos el número de estados (por ejemplo, posiciones en un mapa)
estados = 5

# Definimos el número de acciones posibles (izquierda o derecha)
acciones = 2

# Creamos la tabla Q (donde el agente aprende)
# Inicialmente está llena de ceros
Q = np.zeros((estados, acciones))

# Definimos parámetros del aprendizaje
alpha = 0.1   # tasa de aprendizaje (qué tanto aprende cada vez)
gamma = 0.9   # factor de descuento (importancia del futuro)
epsilon = 0.2 # probabilidad de explorar (probar algo nuevo)

# Definimos el número de episodios (veces que el agente intenta aprender)
episodios = 100

# Definimos la posición objetivo (donde hay recompensa)
objetivo = 4

# Iniciamos el proceso de entrenamiento
for episodio in range(episodios):

    # Empezamos siempre desde el estado 0
    estado = 0

    # Mientras no lleguemos al objetivo
    while estado != objetivo:

        # Elegimos acción: explorar o explotar
        if np.random.rand() < epsilon:
            # Exploración: acción aleatoria
            accion = np.random.randint(acciones)
        else:
            # Explotación: elegimos la mejor acción según la tabla Q
            accion = np.argmax(Q[estado])

        # Aplicamos la acción
        if accion == 0:
            # Acción 0: moverse a la izquierda
            nuevo_estado = max(0, estado - 1)
        else:
            # Acción 1: moverse a la derecha
            nuevo_estado = min(estados - 1, estado + 1)

        # Definimos la recompensa
        if nuevo_estado == objetivo:
            recompensa = 1   # recompensa por llegar al objetivo
        else:
            recompensa = 0   # sin recompensa

        # Actualizamos la tabla Q (aprendizaje)
        Q[estado, accion] = Q[estado, accion] + alpha * (
            recompensa + gamma * np.max(Q[nuevo_estado]) - Q[estado, accion]
        )

        # Pasamos al nuevo estado
        estado = nuevo_estado

# Mostramos la tabla Q final
print("Tabla Q aprendida:")
print(Q)
```

```python
# Importamos numpy para manejar datos numéricos
import numpy as np

# Importamos el modelo LabelPropagation (aprendizaje semi-supervisado)
from sklearn.semi_supervised import LabelPropagation

# Creamos los datos (X)
# Cada fila representa una persona con dos características:
# [edad, dinero gastado]
X = np.array([
    [18, 100],
    [20, 150],
    [22, 130],
    [40, 800],
    [42, 850],
    [45, 900]
])

# Creamos las etiquetas (y)
# -1 significa que no sabemos la etiqueta (dato sin clasificar)
# 0 y 1 son clases conocidas
y = np.array([0, -1, -1, 1, -1, -1])

# Creamos el modelo de aprendizaje semi-supervisado
modelo = LabelPropagation()

# Entrenamos el modelo con los datos
# Aprende usando tanto los datos etiquetados como los no etiquetados
modelo.fit(X, y)

# Obtenemos las etiquetas predichas para todos los datos
predicciones = modelo.transduction_

# Mostramos las etiquetas finales
print("Etiquetas predichas:", predicciones)
```

```python
# Importamos numpy para manejar datos numéricos
import numpy as np

# Importamos las herramientas de redes neuronales de Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Creamos los datos de entrada (X)
# Cada valor representa horas de estudio
X = np.array([[1], [2], [3], [4], [5], [6]])

# Creamos las etiquetas (y)
# 0 = pierde, 1 = aprueba
y = np.array([0, 0, 0, 1, 1, 1])

# Creamos el modelo secuencial (una red neuronal)
modelo = Sequential()

# Agregamos una capa oculta con 4 neuronas y función de activación relu
modelo.add(Dense(4, input_dim=1, activation='relu'))

# Agregamos la capa de salida con 1 neurona (clasificación binaria)
modelo.add(Dense(1, activation='sigmoid'))

# Compilamos el modelo
# optimizer: cómo aprende
# loss: qué tan mal está (error)
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenamos el modelo con los datos
# epochs = número de veces que aprende de los datos
modelo.fit(X, y, epochs=100, verbose=0)

# Creamos un nuevo dato (una persona que estudió 3.5 horas)
nuevo = np.array([[3.5]])

# Hacemos la predicción
prediccion = modelo.predict(nuevo)

# Mostramos el resultado
print("Probabilidad de aprobar:", prediccion[0][0])
```
